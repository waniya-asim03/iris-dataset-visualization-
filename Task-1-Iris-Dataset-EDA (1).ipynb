Step 1: Import Required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: improve plot style
sns.set(style="whitegrid")
```

Step 2: Load the Iris Dataset

```python
# Load Iris dataset from seaborn
iris = sns.load_dataset("iris")

# Display first 5 rows
iris.head()
```

This shows the first five rows of the dataset, helping us understand feature names and values.


 Step 3: Understand Dataset Structure

```python
# Shape of dataset
iris.shape
```

```python
# Column names
iris.columns
```

```python
# Dataset information
iris.info()
```

```python
# Statistical summary
iris.describe()
```

 Step 4: Scatter Plot (Relationship Between Variables)

```python
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="sepal_length",
    y="petal_length",
    hue="species",
    data=iris
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

Step 5: Histogram (Data Distribution)

```python
plt.figure(figsize=(8,5))
plt.hist(iris["sepal_length"], bins=20)
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()

```

 Step 6: Box Plot (Outliers & Spread)

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="species", y="petal_length", data=iris)
plt.title("Petal Length Distribution by Species")
plt.show()
```

Step 7: Conclusion


- The Iris dataset contains 150 records and 5 attributes.
- Petal length and petal width are strong features for distinguishing species.
- Setosa is clearly separable from Versicolor and Virginica.
- Data visualization helps understand patterns before applying machine learning models.
